### Input Parameters

```python
import math

# ============================================================
# Torque–preload calculation for a plain castellated nut
# Thread: 0.6250-18 UNJF
# ============================================================

# ----------------------------
# Input parameters
# ----------------------------

# Nominal thread diameter
d_in = 0.6250                  # inch

# Threads per inch
threads_per_inch = 18.0

# Friction coefficients
mu_thread = 0.14
mu_bearing = 0.14

# Thread flank half-angle for Unified threads
alpha_deg = 30.0               # degrees

# Approximate nut width across flats
nut_width_flat_in = 0.934      # inch

# Target preload
preload_N = 6000.0             # N

# ----------------------------
# Unit conversion
# ----------------------------

inch_to_mm = 25.4

d_mm = d_in * inch_to_mm
d_m = d_mm / 1000.0

pitch_mm = inch_to_mm / threads_per_inch

nut_width_flat_mm = nut_width_flat_in * inch_to_mm

print("INPUTS")
print(f"Nominal diameter:              {d_mm:.3f} mm")
print(f"Thread pitch:                  {pitch_mm:.4f} mm")
print(f"Thread friction coefficient:   {mu_thread:.3f}")
print(f"Bearing friction coefficient:  {mu_bearing:.3f}")
print()
```

### Thread geometry

```python
# Approximate basic pitch diameter for a 60° Unified thread
d2_mm = d_mm - 0.649519 * pitch_mm

# Lead angle for a single-start thread
lambda_rad = math.atan(pitch_mm / (math.pi * d2_mm))
lambda_deg = math.degrees(lambda_rad)

# Equivalent friction angle for a V-thread
alpha_rad = math.radians(alpha_deg)

rho_rad = math.atan(mu_thread / math.cos(alpha_rad))
rho_deg = math.degrees(rho_rad)

print("THREAD GEOMETRY")
print(f"Pitch diameter d2:             {d2_mm:.3f} mm")
print(f"Lead angle lambda:             {lambda_deg:.3f} deg")
print(f"Equivalent friction angle rho: {rho_deg:.3f} deg")
print(f"Effective bearing diameter:    {D_b_mm:.3f} mm")
print()
```

### Effective nut-bearing diameter

```python
# Preliminary approximation:
# mean of nominal thread diameter and nut width across flats
#
# Replace this with the actual effective bearing diameter
# determined from the washer/nut contact geometry when available.
D_b_mm = (d_mm + nut_width_flat_mm) / 2.0

# ----------------------------
# Nut-factor calculation
# ----------------------------

K_thread = (
    d2_mm
    / (2.0 * d_mm)
    * math.tan(lambda_rad + rho_rad)
)

K_bearing = (
    mu_bearing
    * D_b_mm
    / (2.0 * d_mm)
)

K_total = K_thread + K_bearing

print("NUT FACTOR")
print(f"Thread contribution:           {K_thread:.4f}")
print(f"Bearing contribution:          {K_bearing:.4f}")
print(f"Total nut factor K:            {K_total:.4f}")
print()
```

### Torque calculation

```python
T_thread_Nm = K_thread * preload_N * d_m
T_bearing_Nm = K_bearing * preload_N * d_m
T_total_Nm = K_total * preload_N * d_m

print("TORQUE FOR TARGET PRELOAD")
print(f"Target preload:                {preload_N / 1000:.3f} kN")
print(f"Thread torque:                 {T_thread_Nm:.3f} N·m")
print(f"Bearing torque:                {T_bearing_Nm:.3f} N·m")
print(f"Total tightening torque:       {T_total_Nm:.3f} N·m")
print()
```
